2) Define a function max_of_three() that takes three numbers as arguments and returns the largest of them.

```python
def max_of_three(fn,sn,tn):
    if fn > sn:
        if fn > tn:
            return fn
        else:
            return tn
    else:
        if sn > tn:
            return sn
        else:
            return tn
max_of_three(2,3,2)
```

3) Define a function sum() and a function multiply() that sums and multiplies (respectively) all the numbers in a list of numbers. For example, sum([1, 2, 3, 4]) should return 10, and multiply([1, 2, 3, 4]) should return 24.

```python
def sum(ls):
    tot=0
    for i in range(0,len(ls)):
            tot = tot+ls[i]
    return tot

def mul(ls):
    tot=1
    for i in range(0,len(ls)):
        tot =tot*ls[i]
    return tot

print(sum([1,3,4]))
print(mul([2,2,3]))
```

6) Write a program that will calculate the average word length of a text stored in a file (i.e the sum of all the lengths of the word tokens in the text, divided by the number of word tokens).

```python
from __future__ import division # to do floating point division in Python2.7 (not required in python 3.5)
def avg_wl(text):
    sum_of_wl=0
    no_of_word=0
    for word in text.split():
        sum_of_wl = sum_of_wl+len(word)
        no_of_word = no_of_word+1
           
    
    avg = sum_of_wl/no_of_word
    print("total no of words are %d and average length of word is %.2f" %(no_of_word,avg))
        
avg_wl("hi hello python")
```
