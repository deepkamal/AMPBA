**Question 2 :**
    
Write a function that would accept an integer value (say n) and for all numbers from 1 to that integer value, it would create a dictionary whose each element would be of the format (i, i \* i \* i). 
For example, for n = 4, you would create {1:1, 2:8, 3:27, 4:64}

(10 Marks)

***Answer to Q2***

we will write a function that will accept a Integer number

    step 1 - Apply exception handling so error is not caused when invalid number is passed

    step 2 - create a dict which will be filled with return value

    step 2 - run loop from 1 to n , in each iteration update dict with i and i * i * i 
    
    step 4 - return the dict

```python
def cubeDictGen(n):
    
    try:
        retVal = dict()
        for i in range (1,n+1):
            retVal.update({i:i*i*i})
        return retVal
    except:
        return None

cubeDictGen(10)
```

**Question 3 :**

Write a function that receives a size (say n), and two numbers (low and high) and returns a list of n random numbers in the given range. 

(15 Marks)

*** Answer to Q3 ***

we can make use of np.randomn for random number generation

    step 1 - function will accept a integer value for size, and low and high for lower and upper limit of random numbers

    step 2 - np.random.rand will be called to produce random numbers, it produces random values from 0 to 1
    
    step 3 - to adjust the random values so that they become within high and low, we will use mathematical adjustment of low+randomValue * (high - low)

    step 3 - return the generated list 


```python
import numpy as np

def randomNumberList(size,low,high):
    
    return (low+np.random.rand(size)*(high-low))

randomNumberList(50,37,99)
```

***Answer to Q4***

we can use `string.punctuations` to get list of characters which should be replaced
further we can make use of REGEX (`import re`) to clean the given text

    Step 1 - Store the text in a variable, and dictionary in another variable
    
    Step 2 - define a function which accepts TEXT and replaces extra whitespaces, tabs, and punctuations
    
    Step 3 - define a function which splits the given text and counts all the indexes excluding items where `item.strip()==''`
    
    Step 4 - define a function which accepts a list of words, iterates over it to prepare a count of each words
    
    Step 5 - call the functions defined in step 2,3,4 on variable created in step 1 
    
    Step 6 - once the dictionary with word counts is ready, iterate over dict stored in step1 and print the count of each word

```python
givenText='''Why Apple Is In My Retirement Portfolio

Summary

Apple is a rare bird of a stock. It offers shareholders enormous growth, and could potentially become the next great dividend stock. The company is the largest &&&&&& public company in the world, surpassing Exxon Mobil. The raw fundamentals of this company are virtually incomparable. Over the last several years, I have been all over the board when it comes to Apple (AAPL). During its parabolic ------ ----- - - - - - rise in price about 2 years ago, I wrote several articles urging investors to take some chips off the table, and then when it hit some low points, I suggested it might be time to buy the stock, which I did, but not just for my growth portfolio. The stock seemed to have become a value stock with a rather strong + + + + + + dividend, so I placed it in 3 of my portfolios, and have held it ever since in 2 of the 3 portfolios I currently manage. As of several months ago, I added APPL to my newest retirement portfolio, "Buy The Dips Portfolio", or BTDP. The stock is now [ [ [ [ [ going to be held for the very long term in at least 3 out of 4 of my various portfolios, mainly for dividend income investors, but with an eye towards another strong round of growth.

Apple Is More Than A Reborn Growth Stock, It Is Also A Future Dividend Champion

The recent shareholder-friendly moves ] ] ] ] ] ] that Apple has made should ignite even more intense interest in the stock for both dividend-seeking investors, as well as for growth investors.'''

givenWords=[ 'Apple' , 'Google' , 'growth' , 'portfolios' , 'mortgage' ]

'''Question: You are given a string of text. 
Clean the text to remove for any punctuations, extra white spaces, tab characters and count the number of words within the text. 
After this, count the unique number of words in a text and the frequency of occurrence of each word in that text. 
Finally, you are provided a dictionary. Using the dictionary, count the words and frequency of words in the dictionary that occur in text. '''

import re
import string

def textCleaner(text):
    #this function removes extra whitespaces and punctuations
    #below line cascades Regex Substitution, 
    #from inner most call to outer, we are first replacing multiple spaces, 
    #then we are replacing all punctuation characters using string.punctuations, 
    #then we are replacing multiple line breaks with single line
    #finally when all these replacements are done, 
    #we are using string.replace function to replace tab char - I tried using re.sub with '\\t' but for some reason its not working, thus I copied the text gap i could see and replacing it with blank
    return re.sub('\\n\\n',' ',re.sub('|\\'.join(string.punctuation),'',re.sub(' +',' ',text))).replace('      ','')

cleanText = textCleaner(givenText)

#this is our cleaned up text 
print(cleanText)
print('\n\n') #for line gap
words = cleanText.split(' ') #splitting it with spaces to get list of words in cleanText

uniqWords = dict()


for i in words:
    if i.strip()!='':
        if i in uniqWords:
            uniqWords[i]=uniqWords[i]+1
        else:
            uniqWords[i]=1
        
#Now uniqWords dictionary has all unique words as key of dict along with respective count of occurance of each word
print(uniqWords) #the uniqWords has numeric 2 also, not sure if we should clean this or not - if needed we can handle it in above loop
print('\n\n') #for line gap

#Now we are iterating over the givenWords and finding the count of occurance of each given dictionary word in the givenText

print([aWord+" is "+ str(uniqWords[aWord] if aWord in uniqWords else 0 )+" times" for aWord in givenWords])
print('\n\n') #for line gap
```
